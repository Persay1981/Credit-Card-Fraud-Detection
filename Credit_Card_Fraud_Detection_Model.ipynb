```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
creditCardData = pd.read_csv("/content/creditcard.csv")
creditCardData.head()
```

```python
creditCardData.info()
creditCardData.isnull().sum()
```

```python
#distribution of legit and fraudulent transactions
creditCardData['Class'].value_counts()
```

```python
#storing legit and fraudulent transactions independently
legit = creditCardData[creditCardData.Class == 0]
fraud = creditCardData[creditCardData.Class == 1]

print(legit.shape)
print(fraud.shape)
```

```python
print(legit.Amount.describe())
print()
print(fraud.Amount.describe())
```

```python
creditCardData.loc[:,creditCardData.columns != 'Time'].groupby('Class').mean()
```

```python
# Under-Sampling
legit_sample = legit.sample(n=492)

#concatenating legit_sample and fraud
new_dataset = pd.concat([legit_sample,fraud],axis=0)

print(new_dataset.head())
print()
print(new_dataset.tail())
```

```python
new_dataset['Class'].value_counts()
new_dataset.groupby('Class').mean()
```

```python
# Splitting data into features and targets

X = new_dataset.drop('Class',axis=1) #axis = 0 is row and axis = 1 is column
y = new_dataset['Class']

X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2,stratify=y, random_state = 2)
```

```python
# Logistic Regression Model
model = LogisticRegression()

model.fit(X_train,y_train)
```

```python
X_train_prediction = model.predict(X_train)
training_data_accuracy = accuracy_score(X_train_prediction,y_train)
print(training_data_accuracy)
```

```python
X_test_prediction = model.predict(X_test)
testing_data_accuracy = accuracy_score(X_test_prediction,y_test)
print(testing_data_accuracy)
```

```python
import pickle

with open('credit_card_fraud_detection_model.pkl','wb') as f:
  pickle.dump(model,f)
```
